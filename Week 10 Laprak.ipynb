```python
def faktorial(n):                                                         # baris 1
    if n == 0 or n == 1:                                                  # baris 2
        return 1                                                          # baris 3
    else:                                                                 # baris 4
        return n * faktorial(n - 1)                                       # baris 5

try:                                                                      # baris 6
    angka = int(input("Masukkan bilangan: "))                             # baris 7
    if angka < 0:                                                         # baris 8
        print("Faktorial hanya dapat dihitung untuk bilangan positif.")   # baris 9
    else:                                                                 # baris 10
        hasil = faktorial(angka)                                          # baris 11
        print(f"Nilai faktorial dari {angka} adalah {hasil}")             # baris 12
except ValueError:                                                        # baris 13
    print("Masukkan bilangan bulat yang valid.")                          # baris 14
```

```python
import math                                                   # baris 1

def luas_lingkaran():                                         # baris 2
    radius = float(input("Masukkan jari-jari lingkaran: "))   # baris 3
    luas = math.pi * (radius ** 2)                            # baris 4
    print(f"Luas lingkaran adalah: {luas}")                   # baris 5

def luas_persegi():                                           # baris 6
    sisi = float(input("Masukkan panjang sisi persegi: "))    # baris 7
    luas = sisi ** 2                                          # baris 8
    print(f"Luas persegi adalah: {luas}")                     # baris 9

def luas_segitiga():                                          # baris 10
    alas = float(input("Masukkan panjang alas segitiga: "))   # baris 11
    tinggi = float(input("Masukkan tinggi segitiga: "))       # baris 12
    luas = 0.5 * alas * tinggi                                # baris 13
    print(f"Luas segitiga adalah: {luas}")                    # baris 14

while True:                                                   # baris 15
    print("\nPilih perhitungan luas:")                        # baris 16
    print("1. Luas Lingkaran")                                # baris 17
    print("2. Luas Persegi")                                  # baris 18
    print("3. Luas Segitiga")                                 # baris 19
    print("4. Selesai")                                       # baris 20

    pilihan = input("Masukkan pilihan (1/2/3/4): ")           # baris 21

    if pilihan == "1":                                        # baris 22
        luas_lingkaran()                                      # baris 23
    elif pilihan == "2":                                      # baris 24
        luas_persegi()                                        # baris 25
    elif pilihan == "3":                                      # baris 26
        luas_segitiga()                                       # baris 27
    elif pilihan == "4":                                      # baris 28
        print("Program selesai.")                             # baris 29
        break                                                 # baris 30
    else:                                                     # baris 31
        print("Pilihan tidak valid. Silakan coba lagi.")      # baris 32
```
